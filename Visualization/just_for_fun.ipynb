```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# ipython的特殊语法
%matplotlib inline
```

```python
#创建一个等差数列
X = np.linspace(0,2 * np.pi,100) 
Y = np.sin(X)
plt.plot(X,Y,'^')
```

柱状图

```python
y = [5,10,15,20,30]
x = range(len(y))
# 竖着来
plt.bar(x,y)
```

```python
# 横着来
plt.barh(x,y)
```

柱状图——分组

```python
data = np.random.randint(50,size=(3,5))
X = np.arange(5)
plt.bar(X + 0.0 , data[0] , width = 0.25)
plt.bar(X + 0.25 , data[1] , width = 0.25)
plt.bar(X + 0.50 , data[1] , width = 0.25)
```

柱状图——累计分组

```python
plt.bar(X , data[0])
plt.bar(X , data[1] , bottom=data[0])
plt.bar(X , data[2] , bottom=data[0] + data[1])
```

散点图

```python
data = np.random.rand(1000,2)
x = data[:,0]
y = data[:,1]
plt.scatter(x,y)
```

直方图

```python
data = np.random.rand(1000,2)
x = data[:,0]
plt.hist(x,bins=30)
```

线箱图

```python
data = np.random.rand(10,6)
plt.boxplot(data)
```

多图组合

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax1.plot([1,2,3],[2,4,3])

ax2 = fig.add_subplot(2,2,2)
ax2.boxplot(data)

ax3 = fig.add_subplot(2,2,3)
ax3.hist(x,bins=30)

ax4 = fig.add_subplot(2,2,4)
ax4.plot([4,2,1,2],[4,2,1,1])
```
